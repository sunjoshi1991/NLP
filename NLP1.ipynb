```python
### import nltk nlp package
### import sent_tokenize for sentence tokenizing , word_tokenize for word tokenizing
### pos_tag for parts of speech tagging , WordNetLemmatizer for lemmatizer

import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
from nltk import pos_tag, ne_chunk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem.wordnet import WordNetLemmatizer 


stemmer = PorterStemmer()
lem = WordNetLemmatizer()

### read the data set
with open('Mar10_GroupA.txt','r') as f:
    data = f.readlines()
    #print(data)

f = open('test1.txt','a+')
#### tokenizing
for word in data:
    token_words = word_tokenize(word)
    print(token_words)
    #for i in token_words:
        #f.write(i)
        
```

```python
### sentence tokenizing
import re
f = open('hw1_data.txt', 'a+')

for word in data:
    sentences = sent_tokenize(word)
    #print(sentences,end = ' ')
#     for i in sentences:
#         print(i[21:])
    for i in sentences:
        word1 = " ".join(re.findall("[a-zA-Z]+", i))
        word2= word1.replace('PM', '\n',)
        #print(word2)
        f.write(word2)
```

```python
#### parts of speech tagging

for i in data:
    token_words = word_tokenize(i)
    pos = pos_tag(token_words)
    print(pos)
   


```

```python
### stemmimg
for word in data:
    token_words = word_tokenize(word)
    for i in token_words:
        stem_words = stemmer.stem(i)
        print(stem_words,end = ' ')
```

```python
#### lemmatizing
for word in data:
    token_words = word_tokenize(word)
    for token in token_words:
        lemma_words = lem.lemmatize(token)
        print(lemma_words,end = ' ')
```

```python
### name_entity_recognizing

for word in data:
    #print (data)
    named_entity = ne_chunk(pos_tag(word_tokenize(word)))
    print(named_entity)
    #f.write(named_entity)
```
